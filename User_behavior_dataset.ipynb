```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('/content/user_behavior_dataset.csv')

# Print the first few rows of the dataset
print(data.head())

# Get the info of the dataset
print(data.info())

# Get the description of the dataset
print(data.describe())

# Plot the distribution of app usage time
plt.figure(figsize=(10,6))
sns.histplot(data['App Usage Time (min/day)'], bins=50, kde=True)
plt.title('App Usage Time Distribution')
plt.xlabel('App Usage Time (min/day)')
plt.ylabel('Frequency')
plt.show()

# Plot the distribution of screen on time
plt.figure(figsize=(10,6))
sns.histplot(data['Screen On Time (hours/day)'], bins=50, kde=True)
plt.title('Screen On Time Distribution')
plt.xlabel('Screen On Time (hours/day)')
plt.ylabel('Frequency')
plt.show()

# Plot the distribution of data usage
plt.figure(figsize=(10,6))
sns.histplot(data['Data Usage (MB/day)'], bins=50, kde=True)
plt.title('Data Usage Distribution')
plt.xlabel('Data Usage (MB/day)')
plt.ylabel('Frequency')
plt.show()

# Plot the distribution of age
plt.figure(figsize=(10,6))
sns.histplot(data['Age'], bins=50, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Plot the scatter plot of app usage time vs screen on time
plt.figure(figsize=(10,6))
sns.scatterplot(x=data['App Usage Time (min/day)'], y=data['Screen On Time (hours/day)'])
plt.title('App Usage Time vs Screen On Time')
plt.xlabel('App Usage Time (min/day)')
plt.ylabel('Screen On Time (hours/day)')
plt.show()

# Plot the scatter plot of data usage vs age
plt.figure(figsize=(10,6))
sns.scatterplot(x=data['Data Usage (MB/day)'], y=data['Age'], color='orchid')
plt.title('Data Usage vs Age')
plt.xlabel('Data Usage (MB/day)')
plt.ylabel('Age')
plt.show()

# Select only numeric columns
numeric_cols = data.select_dtypes(include=['int64', 'float64']).columns

# Plot the correlation matrix
plt.figure(figsize=(10,8))
sns.heatmap(data[numeric_cols].corr(), annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
plt.show()

# Print the correlation matrix
print(data[numeric_cols].corr())
```
